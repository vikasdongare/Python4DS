# Concatenation

```python
import numpy as np
import pandas as pd
```

```python
df1= pd.DataFrame({'col1':[1,4,7],'col2':[2,5,8],'col3':[3,6,9]})
df1
```

```python
df2= pd.DataFrame({'col1':[11,14,17],'col2':[12,15,18],'col3':[13,16,19]})
df2
```

```python
df3= pd.DataFrame({'col1':[21,24,27],'col2':[22,25,28],'col3':[23,26,29]})
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
a=pd.concat([df1,df2,df3], keys=[1,2,3])
a
```

```python
a.index
```

```python
pd.concat([df1,df2,df3], keys=[1,2,3,4])
```

```python
pd.concat([df1,df2,df3]).reset_index(drop=True)
```

```python
pd.concat([df1,df2,df3], ignore_index=True)
```

```python
# prevent the result from including duplicate index
df10= pd.DataFrame({'col1':[100,400,700],'col2':[200,500,800],'col3':[300,600,900]}, index=[0,4,5])
df10
```

```python
df1
```

```python
pd.concat([df1,df10])
```

```python
pd.concat([df1,df10], verify_integrity= True)
```

```python
xyz=pd.concat([df1,df2,df3], axis=1)
xyz
```

```python
xyz.columns
```

# Merging
like in SQL

```python
df4= pd.DataFrame({'col1':[1,4,7,10],'col2':[12,15,18,20], 'col3':[13,16,19,25]})
df4
```

```python
df5= pd.DataFrame({'col1':[1,4,7,11],'col2':[2,5,8,21], 'col3':[3,6,9,26]})
df5
```

```python
pd.merge(df5, df4, on='col1', how='inner')
```

```python
df5.merge(df4, on='col1')
```

```python
pd.merge(df5,df4, on='col1', suffixes=['_df5','_df4'])
```

```python
pd.merge(df5,df4, on=['col1','col2'])
```

```python
pd.merge(df5, df4, on='col1', how='outer')
```

```python
pd.merge(df5, df4, how='left', on='col1')
```

```python
pd.merge(df5, df4, how='right', on='col1')
```

```python
df4
```

```python
df6= pd.DataFrame({'A':[1,4,7,11],'B':[2,5,8,21],'C':[3,6,9,26]})
df6
```

```python
pd.merge(df4, df6, left_on='col1', right_on='A')
```
